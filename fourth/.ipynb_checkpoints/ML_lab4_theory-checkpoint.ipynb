### Машинное обучение: метод опорных векторов (Support Vector Machine, SVM)

Метод опорных векторов является одним из самых мощных и интересных методов обучения с учителем. 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# use seaborn plotting defaults
import seaborn as sns; sns.set()
```

Метод опорных векторов может использоваться как для классификации, так и для регрессии. SVM является дискриминантным классификатором, т.е. при классификации проводится граница между кластерами данных.

Для начала создадим набор данных

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='spring');
```

Дискриминантный классификатор пытается провести линию (для двумерного пространства) между двумя наборами данных. Но возникает проблема: мы можем провести сразу множество линий, которые будут разбивать наши классы. 

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='spring')

for m, b in [(1, 0.65), (0.5, 1.6), (-0.2, 2.9)]:
    plt.plot(xfit, m * xfit + b, '-k')

plt.xlim(-1, 3.5);
```

В методе опорных векторов эта проблема решается. В нем не просто рисуется линия, но и рассматривается область вокруг нее. Проиллюстрируем это следующим примером:

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='spring')

for m, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none', color='#AAAAAA', alpha=0.4)

plt.xlim(-1, 3.5);
```

SVM пытается максимизировать расстояние между классами. Оптимальной с этой точки зрения в нашем примере будет средняя линия.

### Обучение классификатора

```python
from sklearn.svm import SVC # "Support Vector Classifier"
clf = SVC(kernel='linear')
clf.fit(X, y)
```

Для лучшего представления созданим функцию, которая будет отображать границы разбиения классов.

```python
def plot_svc_decision_function(clf, ax=None):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    x = np.linspace(plt.xlim()[0], plt.xlim()[1], 30)
    y = np.linspace(plt.ylim()[0], plt.ylim()[1], 30)
    Y, X = np.meshgrid(y, x)
    P = np.zeros_like(X)
    for i, xi in enumerate(x):
        for j, yj in enumerate(y):
            P[i, j] = clf.decision_function(np.array([xi, yj]).reshape(1, -1))
    # plot the margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y.reshape(1,-1), s=50, cmap='spring')
plot_svc_decision_function(clf);
```

Стоит обратить внимание, что пунктирные линии касаются пары точек. Эти точки являются опорными. После обучения классификатора в scikit они хранятся в атрибуте support\_vectors\_.

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='spring')
plot_svc_decision_function(clf)
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=200, facecolors='none');
```

Используя возможности Jupyter Notebook можно исследовать, как распределение точек влияет на опорные векторы

```python
from IPython.html.widgets import interact

def plot_svm(N=10):
    X, y = make_blobs(n_samples=200, centers=2,
                      random_state=0, cluster_std=0.60)
    X = X[:N]
    y = y[:N]
    clf = SVC(kernel='linear')
    clf.fit(X, y)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='spring')
    plt.xlim(-1, 4)
    plt.ylim(-1, 6)
    plot_svc_decision_function(clf, plt.gca())
    plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
                s=200, facecolors='none')

interact(plot_svm, N=[10, 200], kernel='linear');
```

Из исследования видно, что на построение векторов влияют только опорные точки. Добавление прочих точек не влияет на результаты классификации.

### Функции ядра

В методе опорных векторов могут использоваться различные функции ядра. Для того, чтобы определить их актуальность, рассмотрим следующий пример, в котором классы линейно не разделимы.

```python
from sklearn.datasets.samples_generator import make_circles
X, y = make_circles(100, factor=.1, noise=.1)

clf = SVC(kernel='linear').fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='spring')
plot_svc_decision_function(clf);
```

Очевидно, что линейно данные классы разделить невозможно. Для решения подобной задачи можно применить радиальную базисную функцию ядра, которая некоторым образом преобразует входные данные. 

```python
r = np.exp(-(X[:, 0] ** 2 + X[:, 1] ** 2))
```

```python
from mpl_toolkits import mplot3d

def plot_3D(elev=30, azim=30):
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:, 0], X[:, 1], r, c=y, s=50, cmap='spring')
    ax.view_init(elev=elev, azim=azim)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('r')

interact(plot_3D, elev=[-90, 90], azip=(-180, 180));
```

Видно, что теперь классы можно линейно разделить. В SVM, конечно же, функции ядра несколько сложнее. Но общий их смысл в том, что все объекты из обучающей выборки с помощью специальных функций отображения вкладываются в пространство большей размерности, в котором элементы уже будут линейно разделимы. И для разбиения классов будет уже строится не линия, а гиперплоскость.

```python
clf = SVC(kernel='rbf')
clf.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='spring')
plot_svc_decision_function(clf)
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=200, facecolors='none');
```
